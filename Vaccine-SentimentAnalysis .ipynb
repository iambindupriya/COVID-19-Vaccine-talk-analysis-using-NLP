```python
#a function to clean the tweet content

def clean_text(text):
    text=re.sub(r'@[A-Za-z0-9]+','',text) #removes @mentions
    text=re.sub(r'#','',text) #removes the hashtag
    text=re.sub(r'RT[\s]+','',text)#removing retweets
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', text, flags=re.MULTILINE) #removing hyperlinks
    text = re.sub(r'http:\/\/.*[\r\n]*', '', text, flags=re.MULTILINE)
    return text

tweets_df2_copy['content']=tweets_df2_copy['content'].apply(clean_text)

for i in range (1,200):
    print(tweets_df2_copy.content[i]+"\n\n")
    i=i+1
    
    
```

```python
#cleaning content 

def clean_content(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

def strip_emoji(text):
    RE_EMOJI = re.compile(u'([\U00002600-\U000027BF])|([\U0001f300-\U0001f64F])|([\U0001f680-\U0001f6FF])')
    return RE_EMOJI.sub(r'', text)
```

```python
tweets_df2_copy['content']=tweets_df2_copy['content'].apply(clean_content)
```

```python
def cc(text):
    text=re.sub('<.*?>', ' ', text)
    text = text.translate(str.maketrans(' ',' ',string.punctuation))
    text = re.sub('[^a-zA-Z]',' ',text)
    text = re.sub("\n"," ",text)
    text = text.lower()
    text=' '.join(text.split())
    return text

```
