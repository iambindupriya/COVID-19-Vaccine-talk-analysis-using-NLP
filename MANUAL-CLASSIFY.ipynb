```python
sample_df['newcontent']=sample_df['content'].str.lower()
```

```python
# write function for removing @user
import re
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i,'',input_txt)
    return input_txt
```

```python
sample_df['newcontent'] = np.vectorize(remove_pattern)(sample_df['newcontent'], '@[\w]*')
#storing modified content column in a new column called new content
```

```python
#removing http and urls from the tweets 
import re
sample_df['newcontent'] = sample_df['newcontent'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])
```

```python
sample_df['newcontent'] = sample_df['newcontent'].str.replace('[^a-zA-Z#]+',' ')
```

```python
# remove short words
sample_df['newcontent'] = sample_df['newcontent'].apply(lambda x: ' '.join([w for w in x.split() if len(w) > 2]))
```

```python
sample_df
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

sample_df['tweet_punct'] = sample_df['newcontent'].apply(lambda x: remove_punct(x))
```

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text

sample_df['tweet_tokenized'] = sample_df['tweet_punct'].apply(lambda x: tokenization(x.lower()))
```

```python
stopword = stopwords.words('english')
custom_stop_word_list=['#','covid','covid-19',' covid ','vaccines' '# covid','covid vaccine', 'vaccine','covid 19 vaccine','coronavirus vaccine', 'covid #']
final_stopword_list = custom_stop_word_list + stopword

print(final_stopword_list)
```

```python
def remove_stopwords(text):
    text = [word for word in text if word not in final_stopword_list]
    return text
    
sample_df['tweet_nonstop'] = sample_df['tweet_tokenized'].apply(lambda x: remove_stopwords(x))
```

```python
def clean_text(text):
    text_lc = "".join([word.lower() for word in text if word not in string.punctuation]) # remove puntuation
    text_rc = re.sub('[0-9]+', '', text_lc)
    tokens = re.split('\W+', text_rc)    # tokenization
    text = [word for word in tokens if word not in final_stopword_list]  # remove stopwords and stemming
    return text
```

```python
sample_df['tweet_nonstop']=sample_df['tweet_nonstop'].apply(lambda x: ' '.join(x))
```

```python
countVectorizer = CountVectorizer(analyzer=clean_text) 
countVector = countVectorizer.fit_transform(sample_df['content'])
print('{} Number of tweets has {} words'.format(countVector.shape[0], countVector.shape[1]))
print(countVectorizer.get_feature_names())
```

```python
import nltk
from nltk.util import ngrams
def extract_ngrams(data, num):
    n_grams = ngrams(nltk.word_tokenize(data), num)
    return [ ' '.join(grams) for grams in n_grams]
    
 
data = ' '.join(sample_df['tweet_nonstop'])
```

```python
sample_df['label']=sample_df['manual_sentiment']
sample_df.loc[sample_df['label'] == "Positive", 'label'] = 1
sample_df.loc[sample_df['label'] == "Negative", 'label'] = 2
sample_df.loc[sample_df['label'] == "Neutral", 'label'] = 0
```

```python
sample_df
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(sample_df['tweet_nonstop'].values.astype('U'))
```

```python
print(X)
```

```python
Y=sample_df['label'].values
Y=Y.astype('int')
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=42)
```

```python
model=MultinomialNB()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```
